```python
# type your code here

num = int(input("Enter the Number: "))

if num < 0:
    print("Negative")
else:
    if num == 0:
        print(1)
    else:
        fact = 1
        i = 1
        while i <= num:
            fact*=i   # fact=fact*i
            i+=1
        print(fact)
```
