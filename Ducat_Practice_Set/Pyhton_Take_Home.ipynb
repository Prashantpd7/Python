```python
# type your code here
fruity = {'Fruits 01' : 'Apple', 'Fruits 02' : 'Orange', 'Fruits 0 3 ': 'Grapes'}

fruity_new = {}

for key, value in fruity.items():
  new_key = key.replace(" ","")
  fruity_new[new_key] = value

print(fruity_new)

```
