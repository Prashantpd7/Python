### Practice Questions

#### Q1. Create a NumPy array of numbers from 1 to 20. Then reshape it into a 4×5 matrix and print:
1. The matrix
2. The shape
3. The data type

```python
arr = np.arange(1,21).reshape(4,5)
print(arr)
```

#### Q2. From the 4×5 matrix you created earlier, do the following:

1. Print the first row
2. Print the last column
3. Print the element at 3rd row, 4th column
4. Print all elements greater than 10

```python
arr[0]
```

```python
arr[:,-1]
```

```python
arr[2,3]
```

```python
arr[arr>10]
```

```python
arr+10
```

```python
arr*2
```

```python
arr/2
```
