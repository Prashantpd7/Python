### Pyplot API

#### Univariate - Numerical

```python
data = {
    "Salary" : [25000,35000,54000,23000,87000,24000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
# Line Plot

plt.plot(df["Salary"])
plt.show()
```

```python
# Line Plot modified 

plt.plot(df["Salary"], color = "red", marker = "o", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
# Histogram

plt.hist(df["Salary"], bins = 5, color = "green")
plt.show()
```

```python
#Box

plt.boxplot(df["Salary"])
plt.show()
```
